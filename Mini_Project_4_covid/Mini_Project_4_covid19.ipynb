#  1] Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

# 2] Loading dataset into pandas dataframe

```python
df=pd.read_csv('covid_vaccine_statewise.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.State.unique()
```

# measures of central tendency

```python
df.describe()
```

```python
ind_df=df.copy()
ind_df=ind_df.loc[df['State'] == 'India']

rest_df=df.copy()
rest_df.drop(df[(df['State'] =='India') ].index, inplace=True)
```

```python
sum_data=rest_df.groupby(['State']).sum().reset_index()
```

```python
sum_data
```

```python
sum_data.columns
```

# visualization

```python
sum_data.plot(kind='bar',x='State',y='Total Doses Administered')
```

```python
plt.rcParams['figure.figsize'] = (25, 7)
ax = sn.barplot(x = sum_data['State'], y =sum_data['Total Doses Administered'], palette = 'dark')
ax.set_xlabel(xlabel = 'Countries', fontsize = 10)
ax.set_ylabel(ylabel = 'Total Doses Administered', fontsize = 10)
ax.set_title(label = 'Total Doses Administered', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```
